# Ajouter une colone 

```python
df['prix_total'] = df['prix'] * (1 - df['tax'] / 100)
```

```python
df
```

Extraction du mois 

```python
def month(x):
    return x.split('-')[1]
```

```python
df['mois'] = df['date_1'].apply(month)
```

```python
df
```

extraction de l'années 

```python
def annees(x):
    return x.split('-')[0]
```

```python
df['annees'] = df['date_1'].apply(annees)
```

```python
df
```

Extraction des cité 

```python
def city(x):
    return x.split(',')[1]
```

```python
df['city'] = df['ville'].apply(city)
```

```python
df
```

Definir les dictionnaires mapping 

```python
city_mapping = {"Atlanta": "AT", "Seattle": "SE", "New York City": "NY"}
```

```python
city_mapping.keys()
```

Verifier les valeurs uniques dans la colonne 'city'

```python
df['city'].unique()
```

Utiliser strip() pour suprimmer les espaces dans la colonne suplémentaire 

```python
df['city'] = df['city'].str.strip()
city_mapping = {keys.strip(): value for keys, value in city_mapping.items()}
```

```python
df["city_code"] = df["city"].map(city_mapping)
```
