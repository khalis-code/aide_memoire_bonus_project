```python
import pandas as pd
import os
import numpy as np
from datetime import datetime
```

```python
dossier_actuel = os.getcwd()
```

```python
chemin_fichier = os.path.join(dossier_actuel, 'data_csv.csv')
```

```python
dataset = pd.read_csv(chemin_fichier, sep=';')
```

```python
dataset_cleaned = dataset.loc[dataset['expedies'].notnull() |
            dataset['vendus'].notnull()] 
dataset_cleaned
```

```python
dataset_cleaned.columns
```

```python
len(dataset_cleaned[dataset_cleaned['col_vides'].notnull()])
```

```python
dataset_cleaned = dataset_cleaned[['jeux', 
                                   'expedies', 
                                   'vendus', 
                                   'facebook', 
                                   'magasin', 
                                   'autres', 
                                   'date_1', 
                                   'prix',
                                   'tax', 
                                   'ville']]
```

```python
dataset_cleaned.columns
```

```python
dataset_cleaned
```

# Renommer les colonnes 

```python
rename_col = dataset_cleaned.rename(columns={
    'jeux' : 'games', 
    'dates_2' : 'date'  
})
```

```python
rename_col
```

```python
rename_col["tax"].fillna(method='bfill', inplace=True)
rename_col["tax"].fillna(method='ffill', inplace=True)
```

```python
rename_col
```

```python
rename_col.dtypes
```

```python
rename_col['prix'] = rename_col['prix'].apply(lambda x : x.replace("$", ""))
```

```python
rename_col
```

```python
for index, row in rename_col.loc[rename_col['vendus'].notna()].iterrows() : 
    vendus = row[2]
    rename_col.loc[index, 'vendus'] = vendus[:-1]
rename_col
```

# Changer les types de donnes 

```python
rename_col['vendus'] = rename_col['vendus'].astype('float64')
rename_col['prix'] = rename_col['prix'].astype('float64')
rename_col['date_1'] = rename_col['date_1'].astype('datetime64[ns]')
```

```python
rename_col.dtypes
```

# Recherche des doublons 

```python
rename_col.duplicated()
```

```python
rename_col.drop_duplicates()
```

# Remplacer les donnees 

```python
df = rename_col.replace(to_replace=np.nan, value=0)
```
