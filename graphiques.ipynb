```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

Utilisation du librairie Matplotlib pour une Graphique en Bar 

```python
categories = ['categorie1', 'categorie2', 'categorie3', 'categorie4', 'categorie5']
values = [10, 25, 15, 30, 20]
```

```python
plt.figure(figsize=(6, 3))
plt.bar(categories, values)
plt.xlabel('categories')
plt.ylabel('values')
plt.title('Bar Graph')
plt.show()
```

Graphique circulaire ou en cammember 

```python
labels = ['categorieA', 'categorieB', 'categorieC', 'categorieD', 'categorieE']
sizes = [10, 25, 15, 30, 20]
```

```python
plt.figure(figsize=(3, 3))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Pie Graph')
plt.show()
```

Graphique en Bar horizontal 

```python
labels = ['Excel', 'PowerBi', 'SQLite', 'Python', 'Stata']
index = [1, 2, 3, 4, 5]
sizes = [10, 25, 15, 30, 20]
```

```python
plt.figure(figsize=(5, 3))
plt.barh(index, sizes, tick_label=labels)
plt.xlabel('Utilisation')
plt.ylabel('Outils d\'Analyses')
plt.show()
```
